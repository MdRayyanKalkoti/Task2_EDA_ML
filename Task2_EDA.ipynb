IMPORT LIBRARIES

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

LOAD DATASET

```python
df=pd.read_csv("datasets/Teen_Mental_Health_dataset.csv")
df.head()
```

CHECK DATASET INFORMATION

```python
df.info()
```

CHECK MISSING VALUES

```python
df.isnull().sum()
```

CHECK COLUMN NAMES

```python
df.columns
```

DAILY SOCIAL MEDIA USAGE DISTRIBUTION

```python
plt.figure(figsize=(8,5))
sns.histplot(df['daily_social_media_hours'], bins=10, kde=True)
plt.title('Distribution of Daily Social Media Usage')
plt.xlabel('Hours')
plt.ylabel('Number of people')
plt.show()
```

STRESS LEVEL DISTRIBUTION

```python
plt.figure(figsize=(8,5))
sns.countplot(x='stress_level', data=df)
plt.title('Distribution of Stress Levels')
plt.xlabel('Stress Level')
plt.ylabel('Number of people')
plt.show()
```

ADDICTION LEVEL DISTRIBUTION

```python
plt.figure(figsize=(8,5))
sns.countplot(x='addiction_level', data=df)
plt.title('Distribution of Addiction Levels')
plt.xlabel('Addiction Level')
plt.ylabel('Number of people')
plt.show()
```

SLEEP HOURS VS STRESS LEVEL

```python
plt.figure(figsize=(8,5))

sns.scatterplot(x='daily_social_media_hours', y='stress_level', data=df)
plt.title('Daily Social Media Hours vs Stress Level')
plt.xlabel('Daily Social Media Hours')
plt.ylabel('Stress Level')
plt.show()
```

ACADEMIC PERFORMANCE  ANALYSIS

```python
plt.figure(figsize=(8,5))
sns.countplot(x='academic_performance', data=df)
plt.title('Distribution of Academic Performance')
plt.xlabel('Academic Performance')
plt.ylabel('Number of people')
plt.show()
```

CORRELATION HEATMAP ANALYSIS

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')   
plt.title('Correlation Heatmap of Mental Health Dataset')
plt.show()
```

FINAL EDA INSIGHTS

Key Insights from the Analyis:

1. Social media usage patterns vary significantly among individuals.
2. Stress and addiction levels show noticeable behavioral trends.
3. Sleep duration appears to influence stress levels.
4. Academic performance distribution highlights varying student outcomes.
5. Correlation analysis reveals relationships between mental health and lifestyle factors.
